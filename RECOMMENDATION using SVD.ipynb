```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('movies.csv')
data_ratings=pd.read_csv('ratings.csv')
```

```python
user_item=data_ratings.pivot(index='userId',columns='movieId',values='rating').fillna(0)
user_item.head()
//user_item[1,:]=0
//user_item.head()
```

```python
Res=user_item.as_matrix()
Res
#users_ids = list(user_item.index)
user_matrix_mean=np.mean(Res,axis=1)
user_matrix_mean
```

```python
R_demeaned = Res - user_matrix_mean.reshape(-1, 1)
R_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 50)
```

```python
sigma.shape
sigma=np.diag(sigma)
#sigma.shape
#Vt.shape
#sigma
```

```python
all_user_predicted_ratings =np.dot(np.dot(U, sigma),Vt)
all_predicted_ratings_normalize=all_user_predicted_ratings+user_matrix_mean.reshape(-1,1)
all_predicted_ratings_normalize
```

```python
pred_df=pd.DataFrame(all_predicted_ratings_normalize,columns=user_item.columns)
pred_df.head()
```

```python
def recommend_movies(predictions_df, userID, movies_df, original_ratings_df, num_recommendations=5):
    
    user_row_number = userID - 1 
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False) 
    print(sorted_user_predictions)
    
    user_data = original_ratings_df[original_ratings_df.userId == (userID)]
    user_full = (user_data.merge(movies_df, how = 'left', left_on = 'movieId', right_on = 'movieId').
                     sort_values(['rating'], ascending=False)
                 )

    print ('User {0} has already rated {1} movies.'.format(userID, user_full.shape[0]))
    print ('Recommending highest {0} predicted ratings movies not already rated.'.format(num_recommendations))
    
    
    recommendations = (movies_df[~movies_df['movieId'].isin(user_full['movieId'])].
         merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
               left_on = 'movieId',
               right_on = 'movieId').
         rename(columns = {user_row_number: 'Predictions'}).
         sort_values('Predictions', ascending = False).
                       iloc[:num_recommendations, :-1]
                      )

    return user_full, recommendations
```

```python
already,predictions =recommend_movies(pred_df, 1, movies, data_ratings, 10)
already
```
